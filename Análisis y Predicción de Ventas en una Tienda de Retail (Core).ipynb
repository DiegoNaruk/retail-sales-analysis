#7 Análisis y Predicción de Ventas en una Tienda de Retail (Core)

#7.1 Transformacion de datos


```python
df["Ventas_por_producto"] = df["Total Amount"] * df["Quantity"]# Creamos una columna nuev para mostrar la cantidad de ventas por producto
df
```

```python
# Luego creamos una columna que nos clasifique las ventas de acuerdo a ventas altas. medias y bajas
df["Clasificacion_ventas"] = df["Ventas_por_producto"].apply(lambda x: "Alta" if x > 1000  else("Media" if x >= 500 else "Baja"))
df
```

# 7.2 Agrupacion y Agragacion

```python
agrupacion = df.groupby("Product Category")["Ventas_por_producto"].sum()
agrupacion
# Aplicamos un group by de acuerdo a categoria de procuto y ventas por producto
```

```python
agregacion = df.groupby(["Product Category", "Date"])["Ventas_por_producto"].agg(["sum", "mean", "max","min","std","var","count"])
agregacion
# Aplicamos gruop by para mostrar las estadisticas descriptiva de cada grupo. Esto evidencia que hay algunos datos que no presentan desviacion estandar ni varianza.
```

```python
#histograma de Age y Ventas por Producto
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribucion de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show
```

```python
# boxplot de total amount
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 2)
plt.boxplot(df['Total Amount'])
plt.title('Boxplot de Total Amount')
plt.ylabel('Total Amount')
plt.show()
#
```
